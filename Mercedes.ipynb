```python
one_hot = pd.get_dummies(df["transmission"]) #one-hot yapmak istediğimiz sütunları buraya ekliyoruz.
one_hot.head()
```

```python
df = pd.concat([df, one_hot], axis= 1) #Birleştirme işlemi yapıyoruz.
df.drop("transmission", axis=1, inplace= True )
df.head() #now our data is all numerical
```

```python
df.corr()["price"]
```

```python
y = df["price"].values
x = df.drop("price", axis=1).values
```

```python
y
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.3, random_state= 42)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
x_train = scaler.fit_transform(x_train)
```

```python
x_test = scaler.transform(x_test)
```

```python
from tensorflow.keras.models import Sequential #model oluşturuyoruz
from tensorflow.keras.layers import Dense #katman oluşturuyoruz
```

```python
x_train.shape
```

```python
model = Sequential()

model.add(Dense(12, activation = "relu"))
model.add(Dense(12, activation = "relu"))
model.add(Dense(12, activation = "relu"))
model.add(Dense(12, activation = "relu"))

model.add(Dense(1))

model.compile(optimizer= "adam", loss= "mse")
```

```python
model.fit(x_train, y_train, validation_data= (x_test,y_test), batch_size = 250 ,  epochs= 300) #test verilerine göre başarıyı otomatik gösterecek.
```

```python
lossDf = pd.DataFrame(model.history.history)
```

```python
lossDf.head()
```

```python
lossDf.plot(); #kayıp ve doğrulama kaybı
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
tahminDizisi = model.predict(x_test)
```

```python
tahminDizisi
```

```python
y_test
```

```python
mean_absolute_error(y_test, tahminDizisi)
```
